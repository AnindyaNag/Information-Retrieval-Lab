```python
docA = "good boy"
docB = "good girl"
docC = "boy girl good"
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
```

```python
response = tfidf.fit_transform([docA,docB,docC])
```

```python
import pandas as pd
pd.DataFrame(response.toarray(),columns = tfidf.get_feature_names())
```
