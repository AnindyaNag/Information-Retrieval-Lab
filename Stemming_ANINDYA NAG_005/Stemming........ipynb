```python
#Stemming
import nltk
nltk.download('punkt')
```

```python
#Porter Stemming.
from nltk.stem import PorterStemmer
```

```python
#Snowball Stemming.
from nltk.stem import SnowballStemmer
```

```python
#Lancaster Stemming.
from nltk.stem import LancasterStemmer
```

```python
#Regexp Stemming.
from nltk.stem import RegexpStemmer
```

```python
#For Porter Stemmer word 
port = PorterStemmer()
word = input('Enter a word to be stemmed: ')
print(word, " : ",port.stem(word))
```

```python
#For Snowball Stemmer word 
snow = SnowballStemmer(language = 'french')
word = input('Enter a word to be stemmed: ')
print(word, " : ",snow.stem(word))
```

```python
# For Lancaster Stemmer word
lan = LancasterStemmer()
word = input('Enter a word to be stemmed: ')
print(word, " : ", lan.stem(word))
```

```python
#For Regexp Stemmer word
reg = RegexpStemmer('ing$|s$|e$|able$', min=4)
word = input('Enter a word to be stemmed: ')
print(word, " : ", reg.stem(word))
```

```python
#tokenization
from nltk.tokenize import word_tokenize
```

```python
#For sentence
#sample = "Information Retrieval refers to the process and procedures of searching & locating and retrieving recorded data and information from a database"
```

```python
sample = input('Enter a sentence to be stemmed: ')
```

```python
# Words tokenization
words = word_tokenize(sample)
words
```

```python
#Porter Stemming.
for w in words:
    print(w, " : ", port.stem(w))
```

```python
#Snowball Stemming.
for w in words:
    print(w, " : ", snow.stem(w))
```

```python
#Lancaster Stemming.
for w in words:
    print(w, " : ", lan.stem(w))
```

```python
#Regexp Stemming.
for w in words:
    print(w, " : ", reg.stem(w))
```

```python
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer, RegexpStemmer

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer(language='english')
regexp = RegexpStemmer('ing$|s$|e$|able$', min=4)

word_list = ["friend", "friendship", "friends", "friendships"]

print("{0:20}{1:20}{2:20}{3:30}{4:40}".format("Word","Porter Stemmer","Snowball Stemmer","Lancaster Stemmer",'Regexp Stemmer'))

for word in word_list:
    print("{0:20}{1:20}{2:20}{3:30}{4:40}".format(word,porter.stem(word),snowball.stem(word),lancaster.stem(word),regexp.stem(word)))
```
