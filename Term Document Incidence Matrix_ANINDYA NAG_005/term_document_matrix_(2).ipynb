```python
import numpy as np
import pandas as pd
d1 = "I love to play cricket".split()
d2 = "I love play cricket and football everyday".split()
d3 = "Dhoni was captain of cricket team".split()
d4 = "Dhoni won the world cup".split()
d5 = "I love a cup of tea everyday".split()
documents = [d1, d2, d3, d4, d5]
word_count = set(d1 + d2 + d3 + d4 + d5)

small_word = []
for i in word_count:
	small_word.append(i.lower())
terms = sorted(small_word)

tdi = np.zeros(shape = (len(terms),len(documents)))
print(tdi)
 

```

```python
for i, t in enumerate(terms):
    for j, d in enumerate(documents):
        if t in d:
            tdi[i][j] = 1




dic = {}
for i in terms:
    for j in range(len(tdi)):
        dic[i] = tdi[j]


for i, j in dic.items():
    print(f"{i}:{j}")

```

```python

df = pd.DataFrame(tdi, index=dic.keys(), columns = ['d1', 'd2', 'd3', 'd4', 'd5'])
df = df.applymap(lambda x: int(x))
print(df)
```
